```python
import pdfplumber

issues_set = [
    "The issues",
    "The appeal",
    "Appeal",
    "The issues on appeal",
    "The claim",
    "The present claim",
    "Background to the dispute",
    "The substantive issue",
    "Issue to be determined",
    "The relevant issues"
]

outcome_set = [
    "Outcome",
    "Conclusion",
    "Decision",
    "Result",
    "Orders",
    "Summary",
    "Judgement",
    "Outcome:"
]
def extract_issues(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text_extracted = ""
        extract_page = False
        set_next_page = False
        for page in pdf.pages:
            text = page
            #Find BOLD text (meaning headers only)
            clean_text = text.filter(lambda obj: obj["object_type"] == "char" and "Bold" in obj["fontname"])
            lst = clean_text.extract_text().split("\n")
            for header in lst:
                if set_next_page:
                    extract_page = False
                    set_next_page = False
                if header in issues_set: #Start extracting only when header is for issues
                    extract_page = True
                if header in outcome_set: #End extraction when header is for outcome
                    set_next_page = True
            if extract_page:
                page_text = text.extract_text()
                text_extracted += page_text + "/n"
        return text_extracted

print(extract_issues("C:/Users/gaoh7/Documents/GitHub/case-outcome-predictor/data/raw/2000_SGCA_1.pdf"))
```
